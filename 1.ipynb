```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

iris= datasets.load_iris()
X= iris.data
Y= iris.target

# print("Features shape:", X[:2])
# print("Labels shape:", Y[:2])

X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2, random_state=42)
scaler= StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)

model= LogisticRegression()
model.fit(X_train, Y_train)
Y_pred= model.predict(X_test)


print("Predictions:", Y_pred)
print("Accuracy: ", accuracy_score(Y_test, Y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(Y_test, Y_pred))
print("\nClassification Report:\n", classification_report(Y_test, Y_pred))
```

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

iris= datasets.load_iris()
X= iris.data
Y= iris.target

# print("Features shape:", X[:2])
# print("Labels shape:", Y[:2])

X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2, random_state=42)
scaler= StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)

model= DecisionTreeClassifier(criterion='entropy', max_depth=3)
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)


print("Predictions:", Y_pred)
print("Accuracy: ", accuracy_score(Y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(Y_test, y_pred))
print("\nClassification Report:\n", classification_report(Y_test, y_pred))
```

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

iris= datasets.load_iris()
X= iris.data
Y= iris.target

# print("Features shape:", X[:2])
# print("Labels shape:", Y[:2])

X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2, random_state=42)
scaler= StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)

model= SVC(kernel='linear')
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)


print("Predictions:", Y_pred)
print("Accuracy: ", accuracy_score(Y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(Y_test, y_pred))
print("\nClassification Report:\n", classification_report(Y_test, y_pred))
```

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

iris= datasets.load_iris()
X= iris.data
Y= iris.target

# print("Features shape:", X[:2])
# print("Labels shape:", Y[:2])

X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2, random_state=42)
scaler= StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)

model= GaussianNB()
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)


print("Predictions:", Y_pred)
print("Accuracy: ", accuracy_score(Y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(Y_test, y_pred))
print("\nClassification Report:\n", classification_report(Y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Let's first run all models to ensure we have the correct predictions and accuracies
# For Logistic Regression
model_lr = LogisticRegression()
model_lr.fit(X_train, Y_train)
y_pred_lr = model_lr.predict(X_test)
accuracy_lr = accuracy_score(Y_test, y_pred_lr)

# For Decision Tree
model_dt = DecisionTreeClassifier(criterion='entropy', max_depth=3)
model_dt.fit(X_train, Y_train)
y_pred_dt = model_dt.predict(X_test)
accuracy_dt = accuracy_score(Y_test, y_pred_dt)

# For Random Forest
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(X_train, Y_train)
y_pred_rf = model_rf.predict(X_test)
accuracy_rf = accuracy_score(Y_test, y_pred_rf)

# For SVM
model_svm = SVC(kernel='linear')
model_svm.fit(X_train, Y_train)
y_pred_svm = model_svm.predict(X_test)
accuracy_svm = accuracy_score(Y_test, y_pred_svm)

# For KNN
model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(X_train, Y_train)
y_pred_knn = model_knn.predict(X_test)
accuracy_knn = accuracy_score(Y_test, y_pred_knn)

# For Naive Bayes
model_nb = GaussianNB()
model_nb.fit(X_train, Y_train)
y_pred_nb = model_nb.predict(X_test)
accuracy_nb = accuracy_score(Y_test, y_pred_nb)

# Create a dictionary of models and their accuracies
models = {
    'Logistic Regression': accuracy_lr,
    'Decision Tree': accuracy_dt,
    'Random Forest': accuracy_rf,
    'SVM': accuracy_svm,
    'KNN': accuracy_knn,
    'Naive Bayes': accuracy_nb
}

# Set up the figure
plt.figure(figsize=(12, 6))

# Create bar chart
model_names = list(models.keys())
accuracies = list(models.values())
bars = plt.bar(model_names, accuracies, color=['blue', 'green', 'red', 'purple', 'orange', 'brown'])

# Add accuracy values on top of the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.4f}',
             ha='center', va='bottom', rotation=0)

# Add labels and title
plt.xlabel('Models', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('Accuracy Comparison of Different Models on Iris Dataset', fontsize=14)
plt.ylim(0, 1.1)  # Setting y-axis limit to ensure space for text
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
